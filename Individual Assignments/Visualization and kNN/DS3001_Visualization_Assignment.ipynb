````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df = pd.read_csv('college_completion.csv')

print(f"Dimensions of the data: {df.shape}")
print(df.head())
````

````python
cross_tab = pd.crosstab(df['control'], df['level'])
print(cross_tab)
````

````python
sns.histplot(df['grad_100_value'], kde = True, bins = 20)
plt.title('Histogram of grad_100_value')
plt.show()

sns.kdeplot(df['grad_100_value'], shade = True)
plt.title('Kernel Density Plot of grad_100_value')
plt.show()

sns.boxplot(x=df['grad_100_value'])
plt.title('Boxplot of grad_100_value')
plt.show()

print(df['grad_100_value'].describe())
````

````python
sns.kdeplot(data = df, x = 'grad_100_value', hue = 'control', fill = True)
plt.title('Grouped Kernel Density Plot of grad_100_value by control')
plt.show()

sns.kdeplot(data = df, x = 'grad_100_value', hue = 'level', fill = True)
plt.title('Grouped Kernel Density Plot of grad_100_value by level')
plt.show()

# It seems that private not-for-profit and 4-year college institutions have the
# best graduation rates because they have higher densities at higher graduation
# percentages
````

````python
df['levelXcontrol'] = df['level'] + ', ' + df['control']

sns.kdeplot(data = df, x = 'grad_100_value', hue = 'levelXcontrol', fill = True)
plt.title('Grouped Kernel Density Plot of grad_100_value by level and control')
plt.show()

# It seems that 4-year private institutions have the best graduation rates as
# their peak density is around 30-50% of students
````

````python
sns.kdeplot(df['aid_value'], fill = True)
plt.title('Kernel Density Plot of aid_value')
plt.show()

sns.kdeplot(data = df, x = 'aid_value', hue = 'level', fill = True)
plt.title('Kernel Density Plot of aid_value grouped by level')
plt.show()

sns.kdeplot(data = df, x = 'aid_value', hue = 'control', fill = True)
plt.title('Kernel Density Plot of aid_value grouped by control')
plt.show()

# The shape differences in the graphs can be attributed to difference in 4-year and
# 2-year colleges, since 2-year colleges are typically cheaper their peak density
# occur at a lower peak financial aid amount compared to 4 year colleges which
# might need to give much more aid since they are more expensive, however less
# students receive it

# Furthermore, public colleges have additional state funding, and tend to be cheaper
# than private colleges so they seem to give out more aid but at a lesser peak amount

# For-profit colleges give a ton of small amounts of aid to students but do not give
# as large of a sum as not-for-profit, since their goal is for the owners to literally
# turn a profit
````

3. The **hazard rate** gives the instantaneous risk of an event occurring, given that it hasn't already happened. It is defined as
$$
h(t) = \dfrac{f(t)}{1-F(t)}
$$
where $F(t)$ is the distribution of arrival times and $f(t)=F'(t)$ is the density of arrival times. Roughly, conditional on $T \ge t$, what is the probability that $T=t$? The cumulative hazard is given by
$$
H(t) = \int_0^t h(z) dz.
$$
The following function computes the hazard rate and cumulative hazard using something called the Nelson-Ahlen estimator. Run your data through it. Interpret the plots for the hazard rate and cumulative hazard rate. What patterns do you see?

```
def hazard(df):
    """ Compute cumulative hazard rate using Nelson-Ahlen estimator. """
    at_risk = df.shape[0]
    arrivals = df['survival'].sort_values().dropna().tolist()
    hazard_rate = []
    times = []
    for t in arrivals:
        select = (df['survival']==t)*(df['Overall Survival Status']=='Deceased')
        deaths_t = df[select].shape[0]
        if deaths_t > 0:
            hazard_rate.append(deaths_t/at_risk)
            times.append(t)
            at_risk -= deaths_t
    cumulative_hazard = np.cumsum(hazard_rate)
    return cumulative_hazard, times, hazard_rate
```
