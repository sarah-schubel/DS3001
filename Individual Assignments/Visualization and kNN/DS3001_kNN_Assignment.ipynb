1. The goal of both regression and classification are to predict a value pased on inputed variables. However, for regression that value is numeric and classification is categorical.

2. A confusion table is used to report true/false positives/negatives in order to evaluate the accuracy of a classification model.

4. Both overfitting and underfitting occur during model training with part of the target data set. Using too little data for training leads to overfitting and using too much of the data to train leads to underfitting. Overfitting means there would be high accuracy on predictions in the training data and lower accuracy on other datasets because it has "memorized" the details that arose during its training which may not be applicable to any other data. Underfitting means there would be low accuracy on preductions in the training data and other datasets as there was not enough data used to allow the model to understand potential underlying patterns.

5.
```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('USA_cars_datasets.csv')
df.head()
```

```python
df = df.drop(columns = ['brand', 'model', 'title_status', 'color', 'vin', 'lot', 'state', 'country', 'condition'])
df = df.dropna()
df.head()
```

```python
maxmin_year = df['year'] = (df['year'] - df['year'].min()) / (df['year'].max() - df['year'].min())
maxmin_mileage = df['mileage'] = (df['mileage'] - df['mileage'].min()) / (df['mileage'].max() - df['mileage'].min())

print(maxmin_year)
print(maxmin_mileage)
```

```python
from sklearn.model_selection import train_test_split

X = df[['year', 'mileage']]
y = df['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

k_values = [3, 10, 25, 50, 100, 300]

mse_list = []

for k in k_values:
    model = KNeighborsRegressor(n_neighbors = k)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    print(f'k = {k}, MSE = {mse}')
    plt.scatter(y_test, y_pred)

    plt.xlabel('Actual Price')
    plt.ylabel('Predicted Price')
    plt.title(f'k = {k}')
    plt.show()

for k, mse in zip(k_values, mse_list):
    print(f'k = {k}, MSE = {mse}')

# As you increase the value of k it seems like the data becomes more and more
# randomized and the MSE value decreases but then increases

# Based on the plot distributions and the MSE values I believe the optimal
# k value for this data is 25 as the plot retains some of its obvious data
# spread but also has the lowest MSE value

# The graphs with lower k values are overfitted because they do not use enough
# surrounding data points which causes the model to seem overly accurate since
# it is using all the details in the points

# The graphs with higher k valyes are underfitted because they use too much
# surrounding data points which prevents it from making accurate predictions
# which is what causes the plot to seem completely randomized
```
